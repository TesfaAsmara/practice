> The 3n + 1 Problem

```python
def process(n):
    # store our generated sequence
    sequence = [n]
    while n > 1:
        # check if even
        n_is_even = n % 2 == 0
        if (n_is_even):
            n = n/2
            sequence.append(n)
        else:
            n = (n*3)+1
            sequence.append(n)
    return sequence

process(22)
```

```python
def equalize_cost(file):
    fin = open(file, "r") 
    fout = open(file.replace(".in", ".out"), "w")

    input = [float(f"{float(line.strip()):2f}") for line in fin.readlines()]

    def read(expenses):
        #str(max(abs(x) for x in [x - y if i != j else -1 for i, x in enumerate(expenses) for j, y in enumerate(expenses)]))
        return sum([x if x > 0 else 0 for x in list(map(lambda x: x - float(f"{sum(expenses)/len(expenses):2f}"), expenses))])
    
    file_index = 0
    output = ""
    while file_index < len(input):
        n = int(input[file_index])
        if n != 0:
            start = file_index+1
            stop = file_index+n+1
            output += f"{read(input[start:stop]):2f}\n"
            file_index = stop
        else:
            break
    fout.write(output)
equalize_cost('1.6.3.in')
```

```python
def display(file):
    fin = open(file, "r")
    fout = open(file.replace(".in", ".out"), "w")

    input = [line.split() for line in fin.readlines()]
    
    def print_lcd(spec):
        s = int(spec[0])
        rows = 2*s+3
        cols = s + 2

        
        
        def top_left_col(arr):
            for i in range(1, 1+s):
                arr[i][0] = "|"
        
        def bot_left_col(arr):
            for i in range(int(rows/2)+1, int(rows/2)+1+s):
                arr[i][0] = "|"
        
        def top_right_col(arr):
            for i in range(1, 1+s):
                arr[i][-1] = "|"
        
        def bot_right_col(arr):
            for i in range(int(rows/2)+1, int(rows/2)+1+s):
                arr[i][-1] = "|"
        
        def top_row(arr):
            for i in range(1, 1+s):
                arr[0][i] = "-"
        
        def mid_row(arr):
            for i in range(1, 1+s):
                arr[int(rows/2)][i] = "-"
        
        def bot_row(arr):
            for i in range(1, 1+s):
                arr[-1][i] = "-"

        # hori = " " + "-"*s + "\n"
        # vertl = " " + "|\n"*s 
        # vertl_forlANDr = " " + "|" + " "*s 
        # vertr = "\n" + " "*s + "|\n" 
        nums = spec[1]
        fit_vert = int(rows/(s+1))
        fit_hori = int(cols/(s+1))
        output = []
        for num in nums:
            lcd = [[" " for col in range(cols)] for row in range(rows)]
            if num == "1":
                top_left_col(lcd)
                bot_left_col(lcd) 
            elif num == "2":
                top_row(lcd)
                top_right_col(lcd)
                mid_row(lcd)
                bot_left_col(lcd)
                bot_row(lcd)
            elif num == "3":
                top_row(lcd)
                top_right_col(lcd)
                mid_row(lcd)
                bot_right_col(lcd)
                bot_row(lcd)
            elif num == "4":
                top_left_col(lcd)
                top_right_col(lcd)
                mid_row(lcd)
                bot_right_col(lcd)
            elif num == "5":
                top_row(lcd)
                top_left_col(lcd)
                mid_row(lcd)
                bot_right_col(lcd)
                bot_row(lcd)
            elif num == "6":
                top_row(lcd)
                top_left_col(lcd)
                mid_row(lcd)
                bot_left_col(lcd)
                bot_right_col(lcd)
                bot_row(lcd)
            elif num == "7":
                top_row(lcd)
                top_right_col(lcd)
                bot_right_col(lcd)
            elif num == "8":
                top_row(lcd)
                top_left_col(lcd)
                top_right_col(lcd)
                mid_row(lcd)
                bot_left_col(lcd)
                bot_right_col(lcd)
                bot_row(lcd)
            elif num == "9":
                top_row(lcd)
                top_left_col(lcd)
                top_right_col(lcd)
                mid_row(lcd)
                bot_right_col(lcd)
                bot_row(lcd)
            elif num == "0":
                top_row(lcd)
                top_left_col(lcd)
                top_right_col(lcd)
                bot_left_col(lcd)
                bot_right_col(lcd)
                bot_row(lcd)

            for i, list in enumerate(lcd):
                lcd[i] = " ".join(list)
            output.append(lcd)
            
        result = ["" for row in range(rows)]
        for i, list in enumerate(output):
            for row in range(rows):
                result[row] += output[i][row]
        for row in result:
            fout.write(f"{row}\n")
        fout.write("\n")
    for item in input:
        print_lcd(item)

display("1.6.4.in")
```

```python

def graphical_editor(file):
    fin = open(file, "r")
    fout = open(file.replace(".in", ".out"), "w")

    input = [line.split() for line in fin.readlines()]

    def find_region_r(x, y, image, region):
        rows = len(image)
        cols = len(image[0])
        cur_pix = image[x][y]
        n_pix = image[x][y+1] if y+1 < cols else ""
        n_cor = [x, y+1] if y+1 < cols else []
        e_pix = image[x+1][y] if x+1 < rows else ""
        e_cor = [x+1, y] if x+1 < rows else []
        s_pix = image[x][y-1] if y-1 > 0 else ""
        s_cor = [x, y-1] if y-1 > 0 else [] 
        w_pix = image[x-1][y] if x-1 > 0 else ""
        w_cor = [x-1, y] if x-1 > 0 else []
        if cur_pix == n_pix and n_cor not in region:
            region.append(n_cor)
            region.append(find_region_r(n_cor[0], n_cor[1], image, region))
        elif cur_pix == e_pix and e_cor not in region:
            region.append(e_cor)
            region.append(find_region_r(e_cor[0], e_cor[1], image, region))
        elif cur_pix == s_pix and s_cor not in region:
            region.append(s_cor)
            region.append(find_region_r(s_cor[0], s_cor[1], image, region))
        elif cur_pix == w_pix and w_cor not in region:
            region.append(w_cor)
            region.append(find_region_r(w_cor[0], w_cor[1], image, region))
        return region

    def parse_command(list, image):
        command  = list[0]
        if command.startswith("I"):
            m, n = command[1], command[2]
            m = int(m)
            n = int(n)
            image = [["O" for col in range(m)] for row in range(n)]
        elif command.startswith("C"):
                for row in image:
                    for col in image:
                        image[row][col] = "O"
        elif command.startswith("L"):
            x, y, c = command[1], command[2], command[3]
            col = int(x)
            row = int(y)
            image[row][col] = c
        elif command.startswith("V"):
            x, y1, y2, c = command[1], command[2], command[3], command[4]
            for row in range(y1, y2+1):
                image[row][x] = "c"
        elif command.startswith("H"):
            x1, x2, y, c = command[1], command[2], command[3], command[4]
            for col in range(x1, x2+1):
                image[y][col] = c
        elif command.startswith("K"):
            x1, y1, x2, y2, c =command[1], command[2], command[3], command[4], command[5]
            for row in range(y1, y2+1):
                for col in range(x1, x2+1):
                    image[row][col] = c        
        elif command.startswith("F"):
            x, y, c = command[1], command[2], command[3]
            col = int(x)
            row = int(y)
            region = [row, col]
            region_cors = find_region_r(row, col, image, region)
            for cor in region_cors:
                row = cor[0]
                col = cor[1] 
                image[row][col] = c
        elif command.startswith("S"):
            name = list[1]
            output = f"{name}\n"
            for i, row in enumerate(image):
                output+= "".join(row) + "\n"
            fout.write(output)
        elif command.startswith("X"):
            quit()
        return image
    image = []
    for command in input:
        image = parse_command(command, image)
            
#graphical_editor("1.6.5.in")
```

```python
def compute_nesw(cors):
    rows = cors[0]
    cols = cors[1]
    return [(rows, cols+1), (rows-1, cols), (rows, cols-1), (rows+1, cols)]

class Image:
    def I(self, m, n):
        self.num_cols = m
        self.num_rows = n
        self.image = [["O" for m in range(self.num_cols)] for n in range(self.num_rows)]

    def C(self):
        for row in range(self.num_rows):
            for col in range(self.num_cols):
                pixel = self.image[row][col]
                if  pixel != "O":
                    pixel = "O"
    
    def L(self, x, y, c):
        col = x
        row = y
        self.image[row][col] = c
    
    def V(self, x, y1, y2, c): # Error
        col = x
        rowStart = y1
        rowEnd = y2+1
        self.image[rowStart:rowEnd][col] = c
    
    def H(self, x1, x2, y, c):
        startCol = x1
        endCol = x2+1
        row = y
        self.image[row][startCol:endCol] = c
    
    def K(self, x1, y1, x2, y2, c):
        startCol = x1
        endCol = x2+1
        startRow = y1
        endRow = y2+1
        self.image[startRow:endRow][startCol:endCol] = c 

    def F(self, x, y, c):
        col = x
        row = y
        region = set((row, col))
        pix_col = self.image[row][col]
        com_side_cors = compute_nesw((row, col))
        for cor in com_side_cors:
            if cor not in region:
                cs_row = cor[0]
                cs_col = cor[1]
                cs_pix_col = self.image[cs_row][cs_col]
                if cs_pix_col == pix_col:
                    region.add(cor)
        
        for cor in region: # Error
            r_row = cor[0]
            r_col = cor[1]
            self.image[r_row][r_col] = c
    
    def S(self, name, file):
        file.write(f"{name}\n")
        for i, row in enumerate(self.image):
            file.write("".join(row) + "\n")
    

    

    

def graphical_editor2(file):
    fin = open(file, "r")
    fout = open(file.replace(".in", ".out"), "w")

    input = [line.split() for line in fin.readlines()]

    for line in input:
        command = line[0]
        if command == "I":
            m = int(line[1])
            n = int(line[2])
            image = Image()
            image.I(m, n)
        elif command == "C":
            image.C()
        elif command == "L":
            x = int(line[1])
            y = int(line[2])
            c = line[3]
            image.L(x, y, c)
        elif command == "V":
            x = int(line[1])
            y1 = int(line[2])
            y2 = int(line[3])
            c = line[4]
            image.V(x, y1, y2, c)
        elif command == "H":
            x1 = int(line[1])
            x2 = int(line[2])
            y = int(line[3])
            c = line[4]
            image.H(x1, x2, y, c)
        elif command == "K":
            x1 = int(line[1])
            y1 = int(line[2])
            x2 = int(line[3])
            y2 = int(line[4])
            c = line[5]
            image.K(x1, y1, x2, y2, c)
        elif command == "F":
            x = int(line[1])
            y = int(line[2])
            c = line[3]
            image.F(x, y, c)
        elif command == "S":
            name = line[1]
            image.S(name, fout)
        elif command == "X":
            break
graphical_editor2("1.6.5.in")
```
